```python
# ==========================================
# Load Data from Phase 1
# ==========================================

import pandas as pd

data = pd.read_csv("stock_prices_clean.csv", index_col=0, parse_dates=True)

print("Data Loaded Successfully")
print("Shape:", data.shape)
data.head()
```

```python
# Step 1: Daily Log Returns

import numpy as np

returns = np.log(data / data.shift(1))

# Drop NaN values (first row will be NaN)
returns = returns.dropna()

print("Step 1 Completed: Daily Log Returns")
print("Shape:", returns.shape)
returns.head()
```

```python
# Step 2: Rolling Volatility (30 Days)

rolling_volatility = returns.rolling(window=30).std()

print("Step 2 Completed: Rolling Volatility")
print("Shape:", rolling_volatility.shape)
rolling_volatility.tail()
```

```python
# Step 3: Correlation Matrix

correlation = returns.corr()

print("Step 3 Completed: Correlation Matrix")
correlation
```

```python
# Step 4: Portfolio Setup

stocks = data.columns.tolist()

# Equal weights
weights = np.array([1 / len(stocks)] * len(stocks))

print("Step 4 Completed: Portfolio Weights")
print("Stocks:", stocks)
print("Weights:", weights)
```

```python
# Step 5: Portfolio Daily Returns

portfolio_returns = returns.dot(weights)

print("Step 5 Completed: Portfolio Returns")
print("Shape:", portfolio_returns.shape)
portfolio_returns.head()
```

```python
# Step 6: Portfolio Volatility

portfolio_volatility = np.sqrt(
    np.dot(weights.T, np.dot(returns.cov(), weights))
)

print("Step 6 Completed: Portfolio Volatility")
print("Portfolio Volatility:", portfolio_volatility)
```

```python
# Step 7: Validation Checks

print("Validation Checks")

print("Total NaNs in Returns:", returns.isnull().sum().sum())
print("Min Correlation:", correlation.min().min())
print("Max Correlation:", correlation.max().max())

print("\nRecent Rolling Volatility:")
print(rolling_volatility.tail())
```

```python
# Step 8: Save Outputs

returns.to_csv("returns.csv")
rolling_volatility.to_csv("rolling_volatility.csv")
correlation.to_csv("correlation_matrix.csv")
portfolio_returns.to_csv("portfolio_returns.csv")

print("Step 8 Completed: Files Saved")
```
