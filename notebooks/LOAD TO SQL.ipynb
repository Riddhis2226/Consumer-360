```python
import pandas as pd
import sqlite3

# File paths
csv_file = r"C:\Users\acer\Downloads\clean_retail.csv"      # your cleaned dataset
db_file = "retailiq.db"              # output SQLite database

# Load CSV
df = pd.read_csv(csv_file)

# Connect to SQLite
conn = sqlite3.connect(db_file)

# Write to SQL table
df.to_sql(
    name="retail_data",
    con=conn,
    if_exists="replace",   # overwrite if exists
    index=False
)

# Close connection
conn.close()

print("✅ CSV successfully converted to SQLite database!")
```
