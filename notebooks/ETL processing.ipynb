```python
import pandas as pd

def run_etl(input_path, output_path):
    print("🚀 Starting ETL...")

    # Load with BOM fix
    df = pd.read_csv(r"C:\Users\acer\Downloads\Online Retail Dataset - UCI.csv", encoding='utf-8-sig')

    # Clean column names
    df.columns = df.columns.str.strip()

    print("Columns:", df.columns.tolist())
    print(f"Initial rows: {df.shape[0]}")

    # Remove null customers
    df = df.dropna(subset=['CustomerID'])

    # Remove cancelled invoices
    df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]

    # Remove invalid values
    df = df[df['Quantity'] > 0]
    df = df[df['UnitPrice'] > 0]

    # Fix datetime
    df['InvoiceDate'] = pd.to_datetime(
        df['InvoiceDate'],
        format="%d-%m-%y %H:%M",
        errors='coerce'
    )

    df = df.dropna(subset=['InvoiceDate'])

    # Feature Engineering
    df['Revenue'] = df['Quantity'] * df['UnitPrice']
    df['YearMonth'] = df['InvoiceDate'].dt.to_period('M').astype(str)
    df['InvoiceDay'] = df['InvoiceDate'].dt.date

    # Fix types
    df['CustomerID'] = df['CustomerID'].astype(int)

    # Save in SAME location (simple)
    df.to_csv(output_path, index=False)

    print(f"✅ ETL completed → {output_path}")


if __name__ == "__main__":
    run_etl(
        input_path="ecommerce-data.csv",
        output_path="clean_retail.csv"
    )
```
